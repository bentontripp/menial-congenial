To begin, we will need to import any additional libraries that will be used. In our case, that's only the SymPy library. Importing a library in python is pretty straight-forwarward. You simply need to use the `import` command. In general, libraries come with multiple functions/methods that can be imported individually. This is typically good practice. But in our case we are just going to import the full functionality of the SymPy library. This is done with by running `from sympy import *`. 

```python
from sympy import *
```

From here, it isn't too difficult. To create variables, you use what are referred to in SymPy as `symbols`. As an example, let's say you want to use the variable $x$ in your equation. First, you must create $x$ as a variable. This is done with the following line of code:

```python
# create a variable "x"
x = symbols('x')
```

The first thing I would like to note is the first line of code, `# create a variable "x"`. This line is not necessary, just used as a "comment" for whomever is reading the code. Any line beginning with `#` in Python is considered a comment. A comment is often used a reference to explain what a section of code is doing.

The second line is the important one: `x = symbols('x')`. If you recall, I mentioned that SymPy uses `symbols` as a function to define variables. In this line, it is declared that $x$ is a symbol, denoted by the `'x'` argument within the `symbol` function's brackets. `'x'` is what is referred to in the "computer world" as a *string*, or a data-type denoted by Unicode characters. If that's confusing to you, no need to worry. In less tech-savvy terms, a string is literally just some data stored in a "string" of characters. Anything within quotes in pyton is considered a string (they can be single or double quotes). 

To recap, to create a variable we need to *declare* what it will be. It doesn't need to be $x$, it can be almost anything. Let's try using $y$ instead:

```python
y = symbols('y')
print(y)
```

This time, I added an extra line of code: `print(y)`. What is this for? Well you might have noticed that after running this cell, it literally "printed" $y$. The `print()` function is used to display whatever argument is put inside of its brackets. In this case, we created the variable (symbol) `y`, and then printed it. 

Now, to simplify things I will show how to create multiple variables in just one line of code:

```python
# create multiple variables
a, b, c = symbols('a b c')
```

Notice that each of the declared variables are separated by a comma, and the variables listed within the string (the parameter for the `symbols()` function) are separated by a space. As long as this syntax is followed (with the variable names and order matching), you can create as many variables as you need!

For a shortcut to create some of the most common symbols used (i.e. $x$, $y$, $z$, $t$, etc.), simply use the `init_session()` function at the beginning of your session. It will `import *`, create common symbols, and call `init_printing()`.

### **Plotting in Sympy**
One of the biggest challenges in math, especially as it becomes more and more complex, is visualizing what is going on. With that being said, I thought it would be helpful to do one more "introductory" article before diving into some real math. Again, some of these things will be expounded upon later on, but I wanted to at least give some background on how to plot so that it I don't have to explain it too deeply later on.

To plot you will be using the `plotting.plot()` function. By default, SymPy plots use the `matplotlib` library - one of them most popular Python plotting libraries - in the backend. But SymPy plotting is set up so that it is easy to integrate (no pun intended) calculus problems. If you are coding in a notebook, I would also recommend using the `%matplotlib inline` [magic](https://ipython.readthedocs.io/en/stable/interactive/tutorial.html#magics-explained) function at the beginning of your script to automatically print the plot in the notebook.

When calling the function, simply put the calculation that you wish to plot into the paranthases. We will also include the `line_color` argument to specify the color:

```python
%matplotlib inline

p1 = plotting.plot(3 + 2*x, line_color='red')
p2 = plotting.plot(2*x**2 - 20, line_color='blue')
```

To combine two plots into one, just use the `extend` method as follows:

```python
p1.extend(p2)
p1.show()
```

3D plotting is a little more complex. Instead of calling `plotting.plot()`, instead call the `plotting.plot3d()` function. In addition to the calculation to be plotted, it is also necessary to specify the $x$ and $y$ limits of the plot. This is done through tuples of the following format: <br>

`(variable, variable minimum, variable maximum)`

```python
p3 = plotting.plot3d(2*x**2 + 3*y**2, (x, -5, 5), (y, -5, 5))
```
